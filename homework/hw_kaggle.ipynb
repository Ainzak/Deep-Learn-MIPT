Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X_origin = data[feature_cols]
#Совершаем One-hot-encoding
dummy_features = pd.get_dummies(data[cat_cols])
X = pd.concat([data[num_cols], dummy_features], axis=1)
#Нормируем числовые признаки
scaler = StandardScaler()
y = data[target_col]
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values,
                                                    train_size=0.8)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LogisticRegressionCV(scoring='roc_auc', refit = True)
model.fit(X_train_scaled, y_train)
model.predict(X_test)
from sklearn.metrics import roc_auc_score

# Предсказания модели на валидационной выборке
y_pred = model.predict(X_test_scaled)

# Вычисляем ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred)
print(roc_auc)
print("Лучшее значение C:", model.C_[0])
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

```python
# YOUR CODE
# !pip install catboost
import catboost
X_train_origin, X_test_origin, y_train, y_test = train_test_split(X_origin.values, y.values,
                                                       train_size=0.8)
boosting_model = catboost.CatBoostClassifier(learning_rate = 0.1, eval_metric='AUC', depth = 6, n_estimators=180)
boosting_model.fit(X_train_origin, y_train, cat_features=[X_origin.columns.get_loc(col) for col in cat_cols])
```

```python
y_train_predicted = boosting_model.predict_proba(X_train_origin)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_test_origin)[:, 1]
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)
print(train_auc, test_auc)
```
